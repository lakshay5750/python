```python
##how we have to make the new data types that is basically fraction class

class Fraction:
  def __init__(self,x,y):
    self.num=x
    self.deno=y

  def __str__(self):
    return '{}/{}'.format(self.num,self.deno)

  def __add__(self ,other):
    sum_num= self.num*self.deno + other.num*self.deno
    sum_deno=self.deno*other.deno
    return '{}/{}'.format(sum_num,sum_deno)
  def __sub__(self,other):
    sub_num=self.num*other.deno-other.num*self.deno
    sub_deno=self.deno*other.deno
    return '{}/{}'.format(sub_num,sub_deno)
  def __mul__(self,other):
    mul_num=self.num*other.num
    mul_deno=self.deno*other.deno
    return '{}/{}'.format(mul_num,mul_deno)

  def __truediv__(self,other):
    div_num=self.num*other.deno
    div_deno=self.deno*other.num
    return '{}/{}'.format(div_num,div_deno)
```

```python
f1=Fraction(3,4)
f2=Fraction(3,4)
```

```python
print(f1+f2)
print(f1-f2)
print(f1*f2)
print(f1/f2)
```
