```python
##atm class build
class Atm:
  def __init__(self):
    self.pin=''
    self.balance=0

  def menu(self):
    user_input=input(''' Hi.How can i help you?
    1. Press 1 to create pin.
    2. Press 2 to change pin.
    3. Press 3 to check balance.
    4  Press 4 to withdrawl balance.
    5. Press 5 to deposit balance.
    ''')

    if user_input=='1':
      ##enter the userpin
      self.create_pin()

    elif user_input=='2':
      self.change_pin()

    elif user_input=='3':
      print(self.check_balance())
      self.menu()

    elif user_input=='4':

      print( self.withdrawl_balance())
      self.menu()

    elif user_input=='5':
      print(self.deposit())
      self.menu()
    else:
      exit

  def create_pin(self):
    first_pin=input("enter the pin")
    self.pin=first_pin
    self.balance=int(input("enter the amount you currently have"))
    print("pin is successufully created!")
    self.menu()


  def change_pin(self):
    old_pin=input("enter your old pin")
    if self.pin==old_pin:
      new_pin=input("enter your new pin")
      self.pin=new_pin
      self.menu()
    else:
      print("please enter the correct pin")
      self.menu()

  def check_balance(self):
    check_pin=input("enter your pin")
    if check_pin==self.pin:
      return f"your current balance is {self.balance}"
    else:
      return f" please enter the correct pin"
  def withdrawl_balance(self):
    check_pin=input("enter the pin")
    if check_pin==self.pin:
      amount=int(input("please enter the amout for withdrawl"))
      if amount>self.balance:
        return f"you have not enough balance "
      else:
        self.balance-=amount
        print("am0unt withdrawl successfully")
        print(f"your current balance is {self.balance}")
    else:
      return f"please enter the correct pin"

  def deposit(self):
    check_pin=input("enter the pin")
    if self.pin==check_pin:
      amount=int(input("enter the amount you want to deposit"))
      self.balance+=amount
      return f"now your current balance is {self.balance}"

    else:
      return f"please enter the correct pin"
```
