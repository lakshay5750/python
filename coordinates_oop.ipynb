```python
class Point:
  def __init__(self,x,y):
    self.x_cord=x
    self.y_cord=y
  def __str__(self):
    return '({},{})'.format(self.x_cord,self.y_cord)

  def origin_dist(self):
    return f"the distance is from origin is {((self.x_cord)**2+(self.y_cord)**2)**0.5}"
  def cord_dist(self,other):
    dist=((self.x_cord-other.x_cord)**2+(self.y_cord-other.y_cord)**2)**0.5
    return f"the distance between two cordinate is {dist}"
class Line:
  def __init__(self,A,B,C):
    self.A=A
    self.B=B
    self.C=C
  def __str__(self):
   return '({})x + ({})y+ ({})'.format(self.A,self.B,self.C)
  def point_on_line(line,point):
    info=line.A*point.x_cord+line.B*point.y_cord+line.C
    if info==0:
      print('point is present on line')
    else:
      print('point is not present on line')
##finding of the shortest distance from the line of a point
  def shortest_dist(line,point):
    dist=abs(line.A*point.x_cord+line.B*point.y_cord+line.C)/((line.A)**2 + (line.B)**2)**0.5
    return dist
```
