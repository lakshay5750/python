```python
##chunking concept : is basically if we have large amount of fiile
##then it print that len of chunk and go to harddisk and the next
##pointer that in file is read the next key word
##example
with open('sample1.txt','r') as f:
  chunk=2
  data=f.read(chunk)
  while(len(data)>0):
    print(data,end='***')
    f.read(chunk)
```

```python
class Person:
  def __init__(self,name,age,gender):
    self.name=name
    self.age=age
    self.gender=gender

  def display(self):
    print(f"name is {self.name}")
    print(f"age is {self.age}")
    print(f"gender is {self.gender}")
```

```python
person=Person('Lakshay',22,'male')
```
